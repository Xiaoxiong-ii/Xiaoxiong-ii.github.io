```python
#Make line chart
plt.figure(figsize=(14, 8))
for sex in ['Male', 'Female']:
    sex_data = data[data['Sex'] == sex]

    #Plot the number of participants at different levels of education
    plt.plot(sex_data['School_year'], sex_data['Pre_Primary (Thousands)'], marker='o', label=f'{sex} Pre-Primary')
    plt.plot(sex_data['School_year'], sex_data['Primary (Thousands)'], marker='o', label=f'{sex} Primary')
    plt.plot(sex_data['School_year'], sex_data['Secondary (Thousands)'], marker='o', label=f'{sex} Secondary')
    plt.plot(sex_data['School_year'], sex_data['Post_Secondary (Thousands)'], marker='o', label=f'{sex} Post-Secondary')

plt.title('Trends in the Number of Participants in Gender Education')
plt.xlabel('School Year')
plt.ylabel('Number of Participants (Thousands)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
